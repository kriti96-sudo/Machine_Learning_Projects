# Collaborating Filtering Based Recommendation system

```python
x = ratings_books.groupby('User-ID').count()['rating'] >200
act_user = x[x].index
```

```python
filter_rating=ratings_books[ratings_books['User-ID'].isin(act_user)]
```

```python
y = filter_rating.groupby('title').count()['rating']>=50
famous_books=y[y].index
```

```python
final_rating=filter_rating[filter_rating['title'].isin(famous_books)]
```

```python
final_rating
```

```python
pt = final_rating.pivot_table(index='title',columns='User-ID',values='rating')
```

```python
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    # fetch index value
    index = np.where(pt.index==book_name)[0][0]
    similar_item = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    for i in similar_item:
        print(pt.index[i[0]])
    #return suggestions
```

```python
recommend('Message in a Bottle')
```

```python
pt.index[500]
```

```python
recommend('The Notebook')
```

```python
recommend('The Catcher in the Rye')
```
